```python
%%writefile code.js
// Part A

const data = [5, 12, 3, 20, 11, 9, 30];

//loop
let result1=[];

for (let i=0;i <data.length;i++){
  if (data[i]>10)
  {
    result1.push(data[i]*2)
  }
}
//filter
let result2=data.filter(x =>x>10).map(x=>x*2);
//map


//compare
console.log(" Part A Loop");
console.log(result1);
console.log(" Part A Filter and Map")
console.log(result2);



// Part B
const numbers = [1, 2, 3, 4, 5, 6];
//filter and double evens
let evens=numbers.filter(x=>x % 2 === 0).map(x=>x*2);
//make sum
let sum=0
for (let i=0;i <evens.length;i++){
  sum=sum+evens[i];
 }
//print
console.log("Part B sum");
console.log(sum);



// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];
//some
console.log("Part C Some");

function agecheck(user){
  if (user.age<18);
  {
    return true
  }
}
console.log(users.some(agecheck));


//find
console.log("Part C Find");
function idcheckfor3(user){
  if (user.id===3){
    console.log(user);
  }
}
users.find(idcheckfor3);

//forEach
console.log("Part C ForEach");
function checkname(user){
  console.log(user.name);
}
users.forEach(checkname);





// Part D
//repeat
function repeat(n, action) {
  // TODO: call action(i) for i = 0..n-1
  for (i=0;i <n;i++)
  {
    action(i)
  }
}
//tests for repeat
repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
